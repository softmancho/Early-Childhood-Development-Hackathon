## 1. Introduction  

### Background

The size of a child’s brain reaches 90 per cent of an adult’s by the age of five. Early childhood is a time of rapid change – particularly for the development of a child’s brain. These early years are critical for lifelong learning and well-being. Life long social and cognitive skills such as talking, reading, playing logical reasoning capacity are developed at this stage. Kids also develop social skills like empathy, friendship and co-operation at this stage of their lives. Asides from their homes. The school environment is the place outside a childs home where they spend the most time. Moreso, children get to spend time with their peers in the school environment creating the ideal opportunity for these skills and bonds to form. The early years are very important in shaping the cognitive and emotional patterns of kids for life, thus influencing learning, relationships and resilience permanently.

However, the importance of this development and learning process, The education of the pre-school child has not been given sufficient attention by the athorities.  Decision-makers continue to treat our early learning and childcare system as a child-minding service for parents returning to work. There is an erroneous notion that kids are too young to be taken seriously with a tendency to take child education as serious business only after the child communication abilities are well established. This is a grave mistake as most times, the opportunity to mould the kid and influence future capacities might have been lost. 

It’s time the value of play-based learning is recognised by making early learning system high-quality, universally accessible and attached to the public education system so that every child can benefit.
